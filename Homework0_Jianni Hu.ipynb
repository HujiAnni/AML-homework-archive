### 1.2 How many different species are there?

```python
#put all the species in a set to remove duplication 
#count the length of the set to find out the number of different species
classes=set(df.iloc[:,4])
n=len(classes)
classes, n
```

To find out how many different species are there, I put the fifth column into a set to remove duplicates, and get the set named classes containing the different species names of the samples. 
###### The length n of the set is 3, which means there are three different species.

### 1.3 How many samples of each species did Anderson record?

```python
#preset the count of each species to be zeros
se,ve,vi=0,0,0
dataset=np.array(df)

#add one to the count if the sample falls in specific species
for s in dataset:
    if s[4]=="Iris-setosa":
        se+=1
    elif s[4]=="Iris-versicolor":
        ve+=1
    else:
        vi+=1

#print out the result indicating the number of samples of each species        
print("There are %d samples of Iris setosa." %se)
print("There are %d samples of Iris versicolor." %ve)
print("There are %d samples of Iris virginica." %vi)
```

##### We can see that Aderson recorded 50 samples pf each species, respectively.

## 3.Visualize this dataset (scatterplot)

To start with, have all the dependent variables ready before the plotting.
Also, given that each species have 50 samples, the colors of the scatter plot could be preset as well.

```python
#get the value
sepal_length=df.iloc[:,0]
sepal_width=df.iloc[:,1]
petal_length=df.iloc[:,2]
petal_width=df.iloc[:,3]

#preset the color for the scatter plot
colors=np.array(['r']*50+['g']*50+['b']*50)
colors
```

##### Here we have the first plot: sepal_length vs. sepal_width.

```python
#sepal_length vs. sepal_width
plt.rcParams['figure.figsize']=[4,4]
plt.scatter(sepal_length, sepal_width,alpha=0.75,color=colors)
plt.xlabel('Sepal_length')
plt.ylabel('Sepal_width')
plt.title("Sepal Length (cm) vs. Sepal Width (cm)")
```

##### The second plot: sepal_length vs. petal_length.

```python
#sepal_length vs. petal_length
plt.rcParams['figure.figsize']=[4,4]
plt.scatter(sepal_length, petal_length,alpha=0.75,color=colors)
plt.xlabel('Sepal_length')
plt.ylabel('Petal_length')
plt.title("Sepal Length (cm) vs. Petal Length (cm)")
```

##### The third plot: sepal_length vs. petal_width.

```python
#sepal_length vs. petal_width
plt.rcParams['figure.figsize']=[4,4]
plt.scatter(sepal_length, petal_width,alpha=0.75,color=colors)
plt.xlabel('Sepal_length')
plt.ylabel('Petal_width')
plt.title("Sepal Length (cm) vs. Petal Width (cm)")
```

##### The fourth plot: sepal_width vs. petal_length.

```python
#sepal_width vs. petal_length
plt.rcParams['figure.figsize']=[4,4]
plt.scatter(sepal_width, petal_length,alpha=0.75,color=colors)
plt.xlabel('Sepal_width')
plt.ylabel('Petal_length')
plt.title("Sepal Width (cm) vs. Petal Length (cm)")
```

##### The fifth plot: sepal_width vs. petal_width.

```python
#sepal_width vs. petal_width
plt.rcParams['figure.figsize']=[4,4]
plt.scatter(sepal_width, petal_width,alpha=0.75,color=colors)
plt.xlabel('Sepal_width')
plt.ylabel('Petal_width')
plt.title("Sepal Width (cm) vs. Petal Width (cm)")
```

##### And the final plot: petal_length vs. petal_width.

```python
plt.rcParams['figure.figsize']=[4,4]
plt.scatter(petal_length, petal_width,alpha=0.75,color=colors)
plt.xlabel('Petal_length')
plt.ylabel('Petal_width')
plt.title("Petal Length (cm) vs. Petal Width (cm)")
```

### Labeling (Getting ready for the scatter plot)

```python
#prepare the label for the 4x4 figures
import itertools as it
arrange=np.vstack(it.product([0,1,2,3],repeat=2))
arrange
```

```python
#mapping the attributes to the label in the 4x4 figures
labels=np.array([[0,'Sepal Length'],[1,'Sepal Width'],[2,'Petal Length'],[3,'Petal Width']])
labels
```

```python
#create the template of the figure
template=[k for i in arrange.ravel() for k in labels if str(i)==k[0]]
template
```

```python
#update the labels as of the attributes
labels=np.vstack(template)[:,1].reshape(16,2)
labels
```
